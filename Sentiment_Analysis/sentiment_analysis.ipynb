```python
import pandas as pd

df = pd.read_csv("sentiment.csv")
df
```

```python
df.isnull().sum()
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z ]', '', text)
    text = re.sub(r'\s+', ' ', text)
    return text

df['clean_text'] = df['text'].apply(clean_text)
df[['text', 'clean_text']]
```

```python
X = df['clean_text']
y = df['sentiment']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')
X_tfidf = tfidf.fit_transform(X)

X_tfidf
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
def predict_sentiment(text):
    text_clean = clean_text(text)
    text_tfidf = tfidf.transform([text_clean])
    return model.predict(text_tfidf)[0]
```

```python
print(predict_sentiment("I really love this"))
print(predict_sentiment("This is horrible"))
print(predict_sentiment("It is okay"))
```

```python
while True:
    user_input = input("Enter text for sentiment analysis (or type 'exit'): ")
    if user_input.lower() == "exit":
        print("Exiting sentiment analysis.")
        break
    print("Predicted Sentiment:", predict_sentiment(user_input))
```

```python
## Observations
- The model predicts sentiment correctly for clearly positive and negative statements.
- Neutral or mixed sentences may sometimes be misclassified.
- Model performance depends on the size and diversity of training data.

```

```python
## Limitations
- The sentiment analysis model is trained on a small dataset, which limits accuracy.
- The system may misclassify sentences with mixed or ambiguous sentiment.
- Context and sarcasm are not handled by the current model.
- The model is limited to text-based input only.
```

```python
## Future Scope
- The dataset can be expanded with more diverse sentiment examples.
- Advanced NLP models such as deep learning or transformers can be used.
- Sarcasm and context-aware sentiment detection can be implemented.
- The system can be deployed as a web or mobile application.
```
